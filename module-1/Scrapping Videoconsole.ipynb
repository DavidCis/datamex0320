```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url="https://es.wikipedia.org/wiki/Videoconsola"
res=requests.get(url)
video=res.text
soup=bs(video,"html.parser")
```

```python
table_generation=soup.find_all("table",{"class":"wikitable"})[0]
table_generation
```

```python
rows=table_generation.find_all("tr")
rows
```

```python
x=[r.text.strip().split()for r in rows]
x
```

```python
import pandas as pd
```

```python
col_names=x[0]
data=x[1:]

df=pd.DataFrame(data, columns=col_names)
df
```
