#### Book name,price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
#your code
resp=req.get(url).content
soup=bs(resp,"lxml")
top=soup.find_all("article")
libros=[]
lib=[]
for i in range(len(top)):
    b=top[i].text.replace("\n","")
    b=top[i].text.strip().split("\n")
    libros.append(b)
libros    
````

````python
for e in libros:
    tmp=[]
    for st in e:
        st = st.strip()
        if st!='':
            tmp.append(st)
    lib.append(tmp)
lib    
````

````python
df = pd.DataFrame(lib,columns = ["Title","Price","Stock","a"])
df=df.drop(["a"],axis=1)
df
````
