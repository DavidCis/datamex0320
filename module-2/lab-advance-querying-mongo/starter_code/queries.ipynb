# Answers
### 1. All the companies that it's name match 'Babelgum'. Retrieve only their `name` field.

```python
list(db.Companies.find({'name':'Babelgum'}))[:5]
```

### 2. All the companies that have more than 5000 employees. Limit the search to 20 companies and sort them by **number of employees**.

```python
list(db.Companies.find({'number_of_employees': {'$gt': 5000}}).limit(20).sort('number_of_employees'))
```

### 3. All the companies founded between 2000 and 2005, both years included. Retrieve only the `name` and `founded_year` fileds.

```python
list(db.Companies.find({'founded_year': {'$gte': 2000, '$lt': 2005}}).limit(5).sort('founded_year'))
```

### 4. All the companies that had a Valuation Amount of more than 100.000.000 and have been founded before 2010. Retrieve only the `name` and `ipo` fields.

```python
list(db.Companies.find({"$and":[{'ipo.valuation_amount':{'$gt':100000000}},
                                {'founded_year':{'$lt':2010}}]},
                      {'name':1,'ipo.valuation_amount':1}).limit(5))
```

### 5. All the companies that have less than 1000 employees and have been founded before 2005. Order them by the number of employees and limit the search to 10 companies.

```python
list(db.Companies.find({"$and":[{'number_of_employees':{'$lt':1000}},
                                {'founded_year':{'$lt':2005}}]}).limit(10).sort('number_of_employees'))
```

### 6. All the companies that don't include the `partners` field.

```python
list(db.Companies.find({'partners':"$not"}))
```

### 7. All the companies that have a null type of value on the `category_code` field.

```python
list(db.Companies.find({'category_code':None}).limit(5))
```

### 8. All the companies that have at least 100 employees but less than 1000. Retrieve only the `name` and `number of employees` fields.

```python
list(db.Companies.find({'number_of_employees':{'$gte':100,"$lt":1000}},
                       {'name':1,'number of employees':1}).limit(10))
                       
```

### 9. Order all the companies by their IPO price descendently.

```python
 list(db.Companies.find({'ipo.valuation_amount':{'$gt':0}}).limit(10).sort('ipo.valuation_amount',-1))
```

### 10. Retrieve the 10 companies with more employees, order by the `number of employees`

```python
list(db.Companies.find().sort('number of employees',-1).limit(10))
```

### 11. All the companies founded on the second semester of the year. Limit your search to 1000 companies.

```python
list(db.Companies.find({'founded_month':{'$gte':6}}).limit(10))
```

### 12. All the companies that have been 'deadpooled' after the third year.

```python
list(db.Companies.find({'deadpooled_year':{'$gte':3}}).limit(10))
```

### 13. All the companies founded before 2000 that have and acquisition amount of more than 10.000.000

```python
list(db.Companies.find({'$and':[{'founded_year':{'$lt':2000}},
                                 {'acquisition.price_amount':{'$gt':10000000}}]}).limit(10))
```

### 14. All the companies that have been acquired after 2015, order by the acquisition amount, and retrieve only their `name` and `acquisiton` field.

```python
 list(db.Companies.find({'acquisition.acquired_year':{'$gte':2015}},
                        {'name':1,'acquisition.acquired_year':1}).limit(10).sort('acquisition.acquired_year'))
```

### 15. Order the companies by their `founded year`, retrieving only their `name` and `founded year`.

```python
list(db.Companies.find({'founded_year':{'$gt':0}},
                       {'name':1,'founded_year':1}).sort('founded_year',-1).limit(10))
```

### 16. All the companies that have been founded on the first seven days of the month, including the seventh. Sort them by their `aquisition price` descendently. Limit the search to 10 documents.

```python
list(db.Companies.find({'founded_day':{'$gte':7}}).sort('acquisitions.price_amount',-1).limit(10))
```

### 17. All the companies on the 'web' `category` that have more than 4000 employees. Sort them by the amount of employees in ascendant order.

```python
list(db.Companies.find({'$and':[{"category_code":'web'},
                       {"number_of_employees":{'$gt':4000}}]}).sort('number_of_employees').limit(10))
```

### 18. All the companies which their acquisition amount is more than 10.000.000, and currency are 'EUR'.


```python
list(db.Companies.find({'$and':[{"acquisition.price_amount":{'gt':10000000}},
                       {"acquisition.price_currency_code":'EUR'}]}).limit(10))
```

### 19. All the companies that have been acquired on the first trimester of the year. Limit the search to 10 companies, and retrieve only their `name` and `acquisition` fields.

```python
list(db.Companies.find({'acquisition.acquired_month':{'$lte':3}},
                       {'name':1,'acquisition':1}).limit(10))
```

### 20. All the companies that have been founded between 2000 and 2010, but have not been acquired before 2011.

```python
list(db.Companies.find({'$and':[{"founded_year":{'$gte':2000}},{'founded_year':{'$lte':2010}},
                       {"acquired_year":{'$gt':2011}}]}).limit(10))
```
